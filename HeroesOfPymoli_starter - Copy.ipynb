## Player Count

```python
Number_of_Players = (purchase_data["SN"].value_counts()).count()
Player_count = pd.DataFrame({"Total Players":[Number_of_Players]})
Player_count
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting

```python
Unique_items = (purchase_data["Item ID"].value_counts()).count()
Average_price = round(purchase_data["Price"].mean(),2)
Purchase_amount = len(purchase_data)
Total_revenue = round(purchase_data["Price"].sum(),2)
Purchasing_analysis = pd.DataFrame({"Number of Unique Items":[Unique_items],"Average Price":[Average_price],"Number Of Purchases":\
                                    [Purchase_amount],"Total Revenue":[Total_revenue]})
Purchasing_analysis
```
